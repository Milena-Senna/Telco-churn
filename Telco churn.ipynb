# varoli categorici trasformati in valori

```python
s = (telco_churn.dtypes == 'object')
telco_object_cols = list(s[s].index)
```

```python
#creo copia per evitare di modificare dati originali
label_telco = telco_churn.copy()

# Applica il codificatore di etichette a ogni colonna con dati categorici
label_encoder = LabelEncoder()
for col in telco_object_cols:
    label_telco[col] = label_encoder.fit_transform(telco_churn[col])
```

```python
label_telco
```

# Splitto

```python
y = label_telco.Churn
X = label_telco.drop('Churn',axis=1,inplace=False)
```

```python
#target
y
```

```python
X
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

# Modellazione e calcolo accuracy

```python
#Ho ritenuto giusto usare il Random Forest Classifier perchè:
#questo modello va bene su dati di tipo diverso, categoriali e non. 
#Una foresta casuale è una raccolta di alberi di decisione casuale che aggrega i dati dei diversi
#alberi decisionali per decidere la classe finale. In ultimo l'ho preferito al decision tree classifier 
#in quanto con quest'ultimo ho notato che l'accuracy veniva minore rispetto a quello che ho usato.
```

```python
# definizione del modello
telco_model = RandomForestClassifier(random_state=1)
# alleno il modello
telco_model.fit(train_X, train_y)
```

```python
val_predictions = telco_model.predict(val_X)
```

```python
#accuracy
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(val_y,val_predictions)
accuracy_telco = accuracy_score(val_y, val_predictions.round())*100
print ('mae:', mae)
print('accuracy:', accuracy_telco)
```
